```python
class Papa:
    def __init__(self, name):
        self.name = name
    def papaname(self):
        return f"Mein {self.name} aur ye mera pariavr hai "

class Mummy:
    def __init__(self, mothername):
        self.mothername = mothername
    def m(self):
        return f"Aur ye {self.mothername} meri patni hai"

class Beta(Papa, Mummy):
    def __init__(self, name, mothername, sonname):
        self.sonname = sonname
        # Call parent class constructors
        Papa.__init__(self, name)
        Mummy.__init__(self, mothername)
    def family(self):
        # Fixed method call from mothername() to m()
        return f"Mere family mein bas mera ek lauta beta hai {self.sonname}"

Parivar = Beta("Sonu", "Chotu ki Mummy", "Chotu")
print(Parivar.family(),Parivar.m(),Parivar.papaname())
```

```python
class Father:
    def __init__(self,voice):
        self.voice= voice
    def speak(self):
        return f"{self.voice} makes a sound"
class Son(Father):
    def __init__(self,name):
        super().__init__("Son")
        self.name= name
    def wow(self):
        return f"{self.name} said wow"
class Daughter(Father):
    def __init__(self,name):
        super().__init__("Daughter")
        self.name= name
    def yee(self):
        return f"{self.name} says yeee"
S= Son("Athrav")
D= Daughter("Vanshika")
print(S.speak(),S.wow(),D.speak(),D.yee())
```

```python
class Sname:
    def __init__(self, name):
        self.name = name
    def n(self):
        # Changed from self.name() to self.name since name is an attribute, not a method
        return f"The name of the student is {self.name} and"

class Students:
    def __init__(self, s_id):
        self.s_id = s_id
    def id(self):
        # Changed from self.s_id() to self.s_id since s_id is an attribute, not a method
        return f"his/her id is {self.s_id} and"

class Sdetail(Sname, Students):
    def __init__(self, name, s_id, details):
        self.details = details
        # Added self as first parameter to parent class constructors
        Sname.__init__(self, name)
        Students.__init__(self, s_id)
    def d(self):
        return f"{self.n()}{self.id()} he/she is good in {self.details}"

D = Sdetail("Aditya", 12516873, "Python")
print(D.n(), D.id(), D.d())
```
